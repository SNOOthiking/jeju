```python
def getLatLng(addr):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+addr
  headers = {"Authorization": "KakaoAK "+key}
  result = json.loads(str(requests.get(url,headers=headers).text))
  match_first = result['documents'][0]['address']
  return float(match_first['y']),float(match_first['x'])
```

```python
coord = [] 
for i in range(len(jeju_df)):
  try:
    coord.append(getLatLng(jeju_df.loc[i]['소재지전체주소']))
  except:
    coord.append(getLatLng(jeju_df.loc[i]['도로명전체주소']))
len(coord)
```

```python
coord1 =list(map(lambda x : str(x).split('(')[1].split(')')[0] , coord))
```

```python
coord_x = list(map(lambda x : str(x).split(',')[0],coord1))
```
