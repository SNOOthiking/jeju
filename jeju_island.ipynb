## 데이터 프레임 두개로 나눈 이유 
- 크롤링 과정에서 오류가 발생 
> 1. 카카오맵에 존재하지만 네이버맵엔 존재하지 않는 경우
> 2. 네이버맵에 존재하지만 카카오맵엔 존재하지 않는 경우 

최대한 많은 데이터를 크롤링하고 싶었지만 한계가 있었다. 더 나은 방법이 있을까 .. 

****
**카카오맵**
> - 카카오맵 특성상 아직 네이버에 비해 데이터가 부족하다. 
> - 리뷰수 30개 이상, 별점은 3.5점 이상을 기준으로 삼았다.
****
**네이버맵**
> - 카카오맵보다 훨씬 많은 데이터를 보유중이다. 하지만 리뷰이벤트 극단적으로 리뷰알바등 보이는 데로 다 믿어서는 안될 네이버맵 리뷰별점 데이터이다.
> - 리뷰수 50개 이상, 별점은 3.7점 이상을 기준으로 삼았다

****
