## 식당이름 + 식당 + 소재지주소 
- 식당을 추가로 붙인 이유는 타 업체와 구분 

약 1300개에서 5백개까지 줄였다. 완료

```python
# 식당 별점 얻기 
b = jeju_df['n_reviews'].copy()
c = jeju_df['n_stars'].copy()
for i in range(630,680):
  print(i)
  index_nz = nz.iloc[i].name
  name = jeju_df.loc[index_nz]['사업장명']
  row_name = jeju_df.loc[index_nz]['소재지전체주소'].split(' ')[2]
  print(f'{name} 식당 {row_name}')
  driver.find_element(by=By.CLASS_NAME,value="input_search").clear() # 검색창 초기화 
  time.sleep(0.5)
  driver.find_element(by=By.CLASS_NAME,value="input_search").send_keys(f'{name} 식당 {row_name}') # 상호명 검색 이 부분이 계속 바뀜.
  driver.find_element(by=By.CLASS_NAME,value="input_search").send_keys(Keys.ENTER) # 엔터누르기   
  time.sleep(5)
  
  try:
    driver.switch_to.frame("entryIframe")
    time.sleep(5)
    if '별점' in driver.find_element(by=By.CLASS_NAME,value='_20Ivz').text:
      stars = driver.find_element(by=By.CLASS_NAME,value='_20Ivz').text.split('\n')[1].split('/')[0]
      reviews = driver.find_element(by=By.CLASS_NAME,value='_20Ivz').text.split(' ')[1].split('블')[0].replace(',','')
      c[index_nz] = float(stars)
      b[index_nz] = int(reviews)
    else : 
      reviews = driver.find_element(by=By.CLASS_NAME,value='_20Ivz').text.split(' ')[1].split('블')[0].replace(',','')
      b[index_nz] = int(reviews)
    print(b[index_nz],c[index_nz])
  except:
    print(jeju_df.loc[index_nz]['사업장명'],b[index_nz])
    pass
  driver.switch_to.default_content()
```

```python
# 특정 이름 카테고리 변경, 사업장 전체이름을 써주면 카테고리 바꿀 수 있음
def change_category(foodname,to_category):
  temp = [] 
  for i in range(len(jeju_df)):
    if foodname in jeju_df.loc[i]['사업장명']: 
      temp.append(i)
  a =jeju_df['업종구분소분류'].copy()
  a[temp] = to_category
  jeju_df['업종구분소분류'] = a 
  return jeju_df
```
